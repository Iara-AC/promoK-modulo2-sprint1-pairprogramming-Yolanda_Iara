```python
energia_nacional = {"tipo_energia":[], "tipo_generacion":[], "fecha":[], "valor": []}

for dato in dict_energia['included']:
    energia_nacional["tipo_energia"].append(dato['type'])
    energia_nacional["tipo_generacion"].append(dato['attributes']['type'])
    energia_nacional["fecha"].append(dato['attributes']['last-update'])
    for value in dato['attributes']['values']:
        energia_nacional["valor"].append(value['value'])

print(energia_nacional)
```

```python
df_energia = pd.DataFrame(energia_nacional)
```
